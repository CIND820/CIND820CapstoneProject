```python
import pandas as pd
path = "/content/drive/MyDrive/CIND 820/Heart_disease_cleveland_new.csv"
data_cleveland = pd.read_csv(path,encoding='utf-8-sig')

data_cleveland[['sex', 'cp','fbs','restecg','exang','slope','thal','target']] = data_cleveland[['sex', 'cp','fbs','restecg','exang','slope','thal','target']].astype(str)

data_cleveland.head()
```

```python
from ydata_profiling import ProfileReport
prof = ProfileReport(data_cleveland)
prof.to_file(output_file='data_cleveland.html')
```

```python
from google.colab import files
files.download("data_cleveland.html")
```

```python
#Identify duplicate rows in Cleveland Clinic Data
cleveland_dups = data_cleveland[data_cleveland.duplicated(keep=False)]
```

```python
#Count number of null rows
if not cleveland_dups.empty:
      cleveland_dup_num = cleveland_dups.shape[0]
else:
    print("No duplicates in dataset")
```

```python
null_cleveland =data_cleveland.isna()
```

```python
data_cleveland[['age','trestbps','chol','thalach','oldpeak','ca']].describe()
```

```python
#Outlier Prediction using z-scores for Cleveland dataset
from scipy import stats

# Create dataframe of integer features
cleveland_integer = data_cleveland[['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca']]

# Calculate z-scores setting outliers to z-scores outside of +/- 3
z_scores = stats.zscore(cleveland_integer)
boundary = 3
outlier_range = (z_scores > boundary) | (z_scores < -boundary)

# Create dataframe of outlier booleans for each record
cleveland_outlier = pd.DataFrame(outlier_range, columns=cleveland_integer.columns)

# Determine if any features for a given instance contain outliers, filter on true and rename columns
cleveland_outlier['outlier_check'] = cleveland_outlier.any(axis=1)
cleveland_outlier_filter = cleveland_outlier[cleveland_outlier['outlier_check']]
cleveland_outlier_filter= cleveland_outlier_filter.rename(columns={'age': 'age_outlier','trestbps': 'trestbps_outlier','chol': 'chol_outlier','thalach': 'thalach_outlier','oldpeak': 'oldpeak_outlier','ca': 'ca_outlier'})

# Join with integer datafrome for complete list
outlier_table= cleveland_integer.merge(cleveland_outlier_filter, left_index=True, right_index=True)
outlier_table = outlier_table[['age','age_outlier','trestbps','trestbps_outlier','chol','chol_outlier','thalach','thalach_outlier','oldpeak','oldpeak_outlier','ca','ca_outlier','outlier_check']]
```
